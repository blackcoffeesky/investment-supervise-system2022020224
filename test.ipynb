```python
import sqlite3

def list_tables_and_columns(db_path):
    try:
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()

        # 获取所有表名（排除 SQLite 内部表）
        cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name NOT LIKE 'sqlite_%' ORDER BY name;")
        tables = cursor.fetchall()

        if not tables:
            print("数据库中没有用户定义的表。")
            return

        for (table_name,) in tables:
            print(f"\n表名: {table_name}")
            print("-" * (len(table_name) + 5))

            # 获取表的字段信息
            cursor.execute(f"PRAGMA table_info('{table_name}')")
            columns = cursor.fetchall()

            # 输出字段信息
            print(f"{'CID':<5}{'字段名':<20}{'类型':<15}{'可空':<6}{'默认值':<15}{'主键':<5}")
            for cid, name, col_type, notnull, dflt_value, pk in columns:
                print(f"{cid:<5}{name:<20}{col_type:<15}{'否' if notnull else '是':<6}{str(dflt_value):<15}{'是' if pk else '否':<5}")

        conn.close()

    except sqlite3.Error as e:
        print(f"数据库错误: {e}")
    except Exception as e:
        print(f"其他错误: {e}")

# 使用示例
if __name__ == "__main__":
    db_file = "stockdata.db"  # 替换为你的 SQLite 数据库文件路径
    list_tables_and_columns(db_file)
```
